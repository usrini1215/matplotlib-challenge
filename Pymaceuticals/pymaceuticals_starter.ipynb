## Observations and Insights 

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np
from scipy.stats import linregress


# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single dataset
merged_df = pd.merge(mouse_metadata,study_results, on="Mouse ID")

# Display the data table for preview
merged_df
```

```python
# Checking the number of mice.
mice_total_all = len(merged_df["Mouse ID"])

print(f"Total number of mice (includes duplicates) : {mice_total_all}")

mice_total_unique = len(merged_df["Mouse ID"].unique())

print(f"Total number of mice (without duplicates) : {mice_total_unique}")
```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 

#this returns only 5
#duplicate_mice_df = merged_df[merged_df[['Mouse ID','Timepoint']].duplicated()]

#gave 12 records
#ids = merged_df["Mouse ID"]
#duplicate_mice_df = merged_df[ids.isin(ids[ids.duplicated()])].sort_values("Mouse ID")

#gave 5 records
#duplicate_mice_df = merged_df[merged_df.duplicated(subset=['Mouse ID','Timepoint'], keep='first')]

duplicate_mice_ids = merged_df.loc[merged_df.duplicated(subset=['Mouse ID','Timepoint']),'Mouse ID'].unique()

duplicate_mice_ids

```

```python
# Optional: Get all the data for the duplicate mouse ID. 

#all_dupe_data_df = merged_df.loc[merged_df.duplicated(subset =["Mouse ID","Timepoint"],keep=False)==True]

all_dupe_data_df = merged_df.loc[merged_df["Mouse ID"].isin(duplicate_mice_ids)]

print(len(all_dupe_data_df))
all_dupe_data_df
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
#clean_df = merged_df.drop_duplicates(subset=['Mouse ID','Timepoint'],keep = "last")

clean_df = merged_df.loc[merged_df["Mouse ID"].isin(duplicate_mice_ids) == False]

clean_df
```

```python
# Checking the number of mice in the clean DataFrame.
mice_total_clean = len(clean_df["Mouse ID"].unique())

print(f"Total number of mice in clean dataframe : {mice_total_clean}")
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# This method is the most straighforward, creating multiple series and putting them all together at the end.

mean_tumor = clean_df.groupby("Drug Regimen")["Tumor Volume (mm3)"].mean()
median_tumor = clean_df.groupby("Drug Regimen")["Tumor Volume (mm3)"].median()
std_tumor = clean_df.groupby("Drug Regimen")["Tumor Volume (mm3)"].std()
var_tumor = clean_df.groupby("Drug Regimen")["Tumor Volume (mm3)"].var()
sem_tumor = clean_df.groupby("Drug Regimen")["Tumor Volume (mm3)"].sem()

tumor_drug_summary_df = pd.DataFrame({"Mean" : mean_tumor,
                                     "Median" : median_tumor,
                                     "Variance" : var_tumor,
                                     "Standard Deviation" : std_tumor,
                                     "SEM" : sem_tumor})

#print(mean_tumor,median_tumor,mode_tumor,std_tumor,var_tumor,sem_tumor)
tumor_drug_summary_df
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin


# Start by getting the last (greatest) timepoint for each mouse
last_point = clean_df.groupby("Mouse ID")[["Timepoint"]].max()
last_point

# Merge this group df with the original dataframe to get the tumor volume at the last timepoint
tumor_df = pd.merge(last_point,clean_df,on=["Mouse ID","Timepoint"])
tumor_df
```

```python
# Put treatments into a list for for loop (and later for plot labels)
#treatments should be ["Capomulin", "Ramicane", "Infubinol", "Ceftamin"]
tr= tumor_df["Drug Regimen"].unique()
treatments = []
for i in tr:
    if i in ("Capomulin", "Ramicane", "Infubinol", "Ceftamin"):
        treatments.append(i)

        
# Create empty list to fill with tumor vol data (for plotting)
tumor_volume_list = []


# Calculate the IQR and quantitatively determine if there are any potential outliers. 
for treatment in treatments:
    treat_tumor = tumor_df.loc[tumor_df['Drug Regimen'] == treatment]
    
    quartiles = treat_tumor["Tumor Volume (mm3)"].quantile([.25,.5,.75])
    lowerq = round(quartiles[0.25],2)
    upperq = round(quartiles[0.75],2)
    iqr = round(upperq-lowerq,2)
    print("----------------------------")
    print(f"Values for drug {treatment}")
    print(f"The lower quartile of Tumor Volume for drug {treatment} is: {lowerq}")
    print(f"The upper quartile of Tumor Volume for drug {treatment} is: {upperq}")
    print(f"The interquartile range of Tumor Volume for drug {treatment} is: {iqr}")
    print(f"The the median of Tumor Volume for drug {treatment} is: {round(quartiles[0.5],2)} ")

    lower_bound = round(lowerq - (1.5*iqr),2)
    upper_bound = round(upperq + (1.5*iqr),2)
    print(f"Values below {lower_bound} could be outliers for drug {treatment}.")
    print(f"Values above {upper_bound} could be outliersfor drug {treatment}.")
    
    
# Locate the rows which contain mice on each drug and get the tumor volumes
all_drugs_rows = tumor_df.loc[tumor_df['Drug Regimen'].isin(treatments)]
    
    
# add subset 
all_drugs_rows = all_drugs_rows[["Mouse ID","Drug Regimen","Tumor Volume (mm3)"]]   
all_drugs_rows

# Determine outliers using upper and lower bounds
# My answer for outliers and boubds was added above in the loop    
```

```python
# Generate a box plot of the final tumor volume of each mouse across four regimens of interest


tumor_volume = all_drugs_rows.groupby('Drug Regimen')['Tumor Volume (mm3)'].apply(list)
tumor_list_df = pd.DataFrame(tumor_volume)
tumor_list_df = tumor_list_df.reindex(treatments)
print(tumor_list_df)

plt.boxplot(tumor_volume, labels=tumor_volume.index)

plt.ylim(0, 90)
plt.show()
```

## Line and Scatter Plots

```python
# Generate a line plot of time point versus tumor volume for a mouse treated with Capomulin
#tumor_df.loc[tumor_df["Drug Regimen"] == "Capomulin"]
#Used above query to find a mouse treated with Capomulin
time_tumor_rec = clean_df[clean_df["Mouse ID"].isin(["u364"])]

time_tumor_data = pd.DataFrame(time_tumor_rec[["Timepoint", "Tumor Volume (mm3)"]])

line_plot_df = pd.DataFrame(time_tumor_data.set_index("Timepoint"))
print(line_plot_df)

lines = line_plot_df.plot.line()
plt.title("Line plot of Timepoint versus tumor volume for mouse u364 using Capomulin")
plt.xlabel('Timepoint')
plt.ylabel('Tumor Volume')
```

```python
# Generate a scatter plot of mouse weight versus average tumor volume for the Capomulin regimen

cap_scatter_full = clean_df[clean_df["Drug Regimen"].isin(["Capomulin"])]
cap_scatter_df = cap_scatter_full[["Mouse ID","Weight (g)", "Tumor Volume (mm3)"]]
cap_scatter_df = cap_scatter_df.reset_index()

cap_scatter_df = cap_scatter_df.sort_values(["Weight (g)"], ascending=True)
cap_grouped_weight = cap_scatter_df.groupby("Weight (g)")["Tumor Volume (mm3)"].mean()

cap_grouped_plot = pd.DataFrame(cap_grouped_weight).reset_index()
print(cap_grouped_weight)
cap_grouped_plot.plot(kind='scatter', x='Weight (g)', y='Tumor Volume (mm3)', grid = True, figsize= (8,8))
plt.title("Scatter plot of mouse weight versus average tumor volume for the Capomulin")
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and linear regression model 
# for mouse weight and average tumor volume for the Capomulin regimen
x_values = cap_grouped_plot["Weight (g)"]
y_values = cap_grouped_plot["Tumor Volume (mm3)"]
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y =" + str(round(slope,2)) + "x + " + str(round(intercept,2))

CorrCoeff = cap_grouped_plot.corr()
CorrCoeff_Val = round(CorrCoeff['Weight (g)'][1],3)
print(f"The correlation coefficient is = {CorrCoeff_Val}")


plt.scatter(x_values, y_values)
plt.plot(x_values,regress_values,"r-")
plt.annotate(line_eq,(20,40),fontsize=10,color="red")
plt.xlabel("Weight")
plt.ylabel("Tumor Volume")
plt.title("Weight Vs. Avg Tumor Vol")
plt.show()

```
